```python
# installing required libraries
!pip install scikit-learn

# importing libraries
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import files
from sklearn.cluster import KMeans

# choose file option
uploaded = files.upload()

# getting uploaded filename
file_name = list(uploaded.keys())[0]

# reading dataset
mall_data = pd.read_csv(file_name)

# displaying first rows
print(mall_data.head())

# checking column names
print(mall_data.columns)

# selecting useful columns
x = mall_data[['Annual Income (k$)', 'Spending Score (1-100)']]

# creating clustering model
kmeans_model = KMeans(
    n_clusters=5,
    random_state=42
)

# fitting the model
mall_data['Cluster'] = kmeans_model.fit_predict(x)

# displaying updated dataset
print(mall_data.head())

# plotting customer clusters
plt.figure(figsize=(8,6))

plt.scatter(
    mall_data['Annual Income (k$)'],
    mall_data['Spending Score (1-100)'],
    c=mall_data['Cluster']
)

plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation Analysis")

plt.show()
```
